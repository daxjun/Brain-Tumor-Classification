```python
# ============================================================
# Brain Tumor MRI Classification — 4-class + Binary
# Dataset: masoudnickparvar/brain-tumor-mri-dataset (~7,000 imgs)
# ============================================================

# ── 0. Setup ──────────────────────────────────────────────────
!pip install -q kaggle pytorch-gradcam

import os
os.environ['KAGGLE_TOKEN'] = 'KGAT_8a26e3be4d21040099b4bfb62adc3357'  # paste your token

# ── 1. Download Dataset ───────────────────────────────────────
!kaggle datasets download -d masoudnickparvar/brain-tumor-mri-dataset -p /content/data --unzip

# ── 2. Imports ────────────────────────────────────────────────
import numpy as np
import matplotlib.pyplot as plt
import random
from pathlib import Path
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, WeightedRandomSampler
from torchvision import datasets, transforms, models
from torchvision.utils import make_grid
from sklearn.metrics import (classification_report, confusion_matrix,
                             ConfusionMatrixDisplay, roc_auc_score,
                             RocCurveDisplay, roc_curve)

SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

# ── 3. Check Structure ────────────────────────────────────────
data_root = Path("/content/data")
for p in sorted(data_root.rglob("*")):
    if p.is_dir(): print(p)

TRAIN_DIR = data_root / "Training"
TEST_DIR  = data_root / "Testing"

# ── 4. Transforms ─────────────────────────────────────────────
IMG_SIZE = 224

train_tf = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.3, contrast=0.3),
    transforms.RandomAffine(degrees=0, shear=10, translate=(0.1, 0.1)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
])

val_tf = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
])

# ── 5. Datasets ───────────────────────────────────────────────
train_ds = datasets.ImageFolder(str(TRAIN_DIR), transform=train_tf)
test_ds  = datasets.ImageFolder(str(TEST_DIR),  transform=val_tf)

# Split test → val + test
n_test = len(test_ds)
n_val  = n_test // 2
n_te   = n_test - n_val
val_ds, test_ds = torch.utils.data.random_split(
    test_ds, [n_val, n_te],
    generator=torch.Generator().manual_seed(SEED)
)

print(f"\nClasses : {train_ds.classes}")
print(f"Train   : {len(train_ds)}")
print(f"Val     : {len(val_ds)}")
print(f"Test    : {len(test_ds)}")

# ── 6. Class Imbalance — WeightedRandomSampler ────────────────
class_counts = np.array([len(list((TRAIN_DIR / c).iterdir()))
                          for c in train_ds.classes])
class_weights = 1.0 / class_counts
sample_weights = [class_weights[label] for _, label in train_ds.samples]
sampler = WeightedRandomSampler(sample_weights, len(sample_weights), replacement=True)

train_loader = DataLoader(train_ds, batch_size=32, sampler=sampler, num_workers=2)
val_loader   = DataLoader(val_ds,   batch_size=32, shuffle=False,   num_workers=2)
test_loader  = DataLoader(test_ds,  batch_size=32, shuffle=False,   num_workers=2)

# ── 7. Class Distribution Plot ────────────────────────────────
plt.figure(figsize=(7, 4))
plt.bar(train_ds.classes, class_counts, color=["#4C72B0","#DD8452","#55A868","#C44E52"])
plt.title("Training Class Distribution"); plt.ylabel("Count")
for i, v in enumerate(class_counts):
    plt.text(i, v + 10, str(v), ha="center", fontweight="bold")
plt.tight_layout(); plt.show()

# ── 8. Visualise Sample Batch ─────────────────────────────────
def denorm(t):
    mean = torch.tensor([0.485, 0.456, 0.406]).view(3,1,1)
    std  = torch.tensor([0.229, 0.224, 0.225]).view(3,1,1)
    return (t * std + mean).clamp(0, 1)

imgs, labels = next(iter(train_loader))
grid = make_grid(denorm(imgs[:16]), nrow=4)
plt.figure(figsize=(10, 6))
plt.imshow(grid.permute(1,2,0)); plt.axis("off")
plt.title("Sample Training Batch — " + " | ".join(
    train_ds.classes[l] for l in labels[:16].tolist()))
plt.show()

# ── 9. Model ──────────────────────────────────────────────────
NUM_CLASSES = len(train_ds.classes)   # 4

model = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)

# Unfreeze last 2 blocks + classifier from the start (dataset is large enough)
for name, p in model.named_parameters():
    p.requires_grad = any(k in name for k in
                          ["features.6", "features.7", "features.8", "classifier"])

in_features = model.classifier[1].in_features
model.classifier = nn.Sequential(
    nn.Dropout(0.4),
    nn.Linear(in_features, 256),
    nn.ReLU(),
    nn.Dropout(0.3),
    nn.Linear(256, NUM_CLASSES),
)
model = model.to(device)

# ── 10. Loss & Optimiser ──────────────────────────────────────
# Class weights for loss too (double protection against imbalance)
cw = torch.tensor(class_weights / class_weights.sum() * NUM_CLASSES,
                  dtype=torch.float).to(device)
criterion = nn.CrossEntropyLoss(weight=cw)
optimizer = optim.AdamW(
    filter(lambda p: p.requires_grad, model.parameters()),
    lr=1e-3, weight_decay=1e-4
)
scheduler = optim.lr_scheduler.OneCycleLR(
    optimizer, max_lr=1e-3,
    steps_per_epoch=len(train_loader), epochs=25
)

# ── 11. Training Loop ─────────────────────────────────────────
def run_epoch(loader, training=True):
    model.train() if training else model.eval()
    total_loss, correct, n = 0.0, 0, 0
    with (torch.enable_grad() if training else torch.no_grad()):
        for imgs, labels in loader:
            imgs, labels = imgs.to(device), labels.to(device)
            logits = model(imgs)
            loss   = criterion(logits, labels)
            if training:
                optimizer.zero_grad(); loss.backward()
                optimizer.step(); scheduler.step()
            total_loss += loss.item() * len(imgs)
            correct    += (logits.argmax(1) == labels).sum().item()
            n          += len(imgs)
    return total_loss / n, correct / n

EPOCHS   = 25
best_val = 0.0
history  = {"train_loss":[], "val_loss":[], "train_acc":[], "val_acc":[]}

for epoch in range(1, EPOCHS + 1):
    tr_loss, tr_acc = run_epoch(train_loader, training=True)
    vl_loss, vl_acc = run_epoch(val_loader,   training=False)

    history["train_loss"].append(tr_loss); history["val_loss"].append(vl_loss)
    history["train_acc"].append(tr_acc);   history["val_acc"].append(vl_acc)

    flag = ""
    if vl_acc > best_val:
        best_val = vl_acc
        torch.save(model.state_dict(), "best_model.pth")
        flag = "  ← saved"

    print(f"Epoch {epoch:02d}/{EPOCHS}  "
          f"Train {tr_loss:.4f}/{tr_acc:.4f}  "
          f"Val {vl_loss:.4f}/{vl_acc:.4f}{flag}")

# ── 12. Fine-tune: Unfreeze All ───────────────────────────────
print("\n=== Fine-tuning full network ===")
for p in model.parameters():
    p.requires_grad = True

optimizer = optim.AdamW(model.parameters(), lr=2e-5, weight_decay=1e-4)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10)

for epoch in range(1, 11):
    tr_loss, tr_acc = run_epoch(train_loader, training=True)
    vl_loss, vl_acc = run_epoch(val_loader,   training=False)
    scheduler.step()

    flag = ""
    if vl_acc > best_val:
        best_val = vl_acc
        torch.save(model.state_dict(), "best_model.pth")
        flag = "  ← saved"

    print(f"FT {epoch:02d}/10  "
          f"Train {tr_loss:.4f}/{tr_acc:.4f}  "
          f"Val {vl_loss:.4f}/{vl_acc:.4f}{flag}")

# ── 13. Training Curves ───────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for ax, keys, title in zip(axes,
        [("train_loss","val_loss"), ("train_acc","val_acc")],
        ["Loss", "Accuracy"]):
    for k, lbl in zip(keys, ["Train","Val"]):
        ax.plot(history[k], label=lbl)
    ax.set_title(title); ax.set_xlabel("Epoch")
    ax.legend(); ax.grid(True)
plt.tight_layout(); plt.show()

# ── 14. Evaluate on Test Set ──────────────────────────────────
model.load_state_dict(torch.load("best_model.pth"))
model.eval()

all_preds, all_labels, all_probs = [], [], []
with torch.no_grad():
    for imgs, labels in test_loader:
        logits = model(imgs.to(device))
        probs  = torch.softmax(logits, dim=1).cpu()
        preds  = probs.argmax(1)
        all_probs.extend(probs.tolist())
        all_preds.extend(preds.tolist())
        all_labels.extend(labels.tolist())

all_probs  = np.array(all_probs)
all_preds  = np.array(all_preds)
all_labels = np.array(all_labels)

print("\n── Classification Report ──")
print(classification_report(all_labels, all_preds,
                             target_names=train_ds.classes))

# Confusion matrix
cm = confusion_matrix(all_labels, all_preds)
disp = ConfusionMatrixDisplay(cm, display_labels=train_ds.classes)
fig, ax = plt.subplots(figsize=(7, 6))
disp.plot(ax=ax, cmap="Blues", colorbar=False)
plt.title("Confusion Matrix — Test Set"); plt.tight_layout(); plt.show()

# ROC-AUC (one-vs-rest)
auc = roc_auc_score(all_labels, all_probs, multi_class="ovr", average="macro")
print(f"\nMacro ROC-AUC: {auc:.4f}")

# ── 15. Binary view (tumor vs no tumor) ───────────────────────
# notumor class index
no_tumor_idx   = train_ds.class_to_idx["notumor"]
binary_labels  = (all_labels != no_tumor_idx).astype(int)
binary_probs   = 1 - all_probs[:, no_tumor_idx]   # P(tumor)
binary_preds   = (binary_probs >= 0.5).astype(int)

print("\n── Binary Report (Tumor / No Tumor) ──")
print(classification_report(binary_labels, binary_preds,
                             target_names=["No Tumor", "Tumor"]))

fpr, tpr, thresholds = roc_curve(binary_labels, binary_probs)
binary_auc = roc_auc_score(binary_labels, binary_probs)
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"AUC = {binary_auc:.4f}", color="#4C72B0", lw=2)
plt.plot([0,1],[0,1],"--", color="gray")
plt.xlabel("FPR"); plt.ylabel("TPR")
plt.title("ROC Curve — Binary (Tumor vs No Tumor)")
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()

# Optimal threshold (max Youden's J)
J = tpr - fpr
best_thresh = thresholds[np.argmax(J)]
print(f"\nOptimal threshold: {best_thresh:.4f}")
binary_preds_opt = (binary_probs >= best_thresh).astype(int)
print(classification_report(binary_labels, binary_preds_opt,
                             target_names=["No Tumor", "Tumor"]))

# ── 16. Grad-CAM Visualization ────────────────────────────────
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget

target_layer = [model.features[-1]]
cam = GradCAM(model=model, target_layers=target_layer)

def show_gradcam(img_path, true_label=None):
    img_pil = Image.open(img_path).convert("RGB").resize((IMG_SIZE, IMG_SIZE))
    img_np  = np.array(img_pil) / 255.0
    tensor  = val_tf(img_pil).unsqueeze(0).to(device)

    with torch.no_grad():
        logits = model(tensor)
        probs  = torch.softmax(logits, dim=1)[0]
        pred   = probs.argmax().item()

    grayscale_cam = cam(input_tensor=tensor,
                        targets=[ClassifierOutputTarget(pred)])[0]
    overlay = show_cam_on_image(img_np.astype(np.float32), grayscale_cam, use_rgb=True)

    fig, axes = plt.subplots(1, 2, figsize=(8, 4))
    axes[0].imshow(img_pil); axes[0].set_title("Original"); axes[0].axis("off")
    axes[1].imshow(overlay); axes[1].axis("off")
    axes[1].set_title(f"Grad-CAM\nPred: {train_ds.classes[pred]} "
                      f"({probs[pred]:.2%})"
                      + (f"\nTrue: {true_label}" if true_label else ""))
    plt.tight_layout(); plt.show()

# Run on a few test images
test_folder = TEST_DIR
sample_imgs = list(test_folder.rglob("*.jpg"))[:6]
for img_path in sample_imgs:
    true_cls = img_path.parent.name
    show_gradcam(img_path, true_label=true_cls)

# ── 17. Single Image Inference ────────────────────────────────
def predict(img_path, threshold=best_thresh):
    img = Image.open(img_path).convert("RGB")
    tensor = val_tf(img).unsqueeze(0).to(device)
    with torch.no_grad():
        probs = torch.softmax(model(tensor), dim=1)[0].cpu().numpy()
    pred_class = train_ds.classes[probs.argmax()]
    tumor_prob = 1 - probs[train_ds.class_to_idx["notumor"]]
    binary     = "Tumor detected ✓" if tumor_prob >= threshold else "No tumor ✗"

    plt.figure(figsize=(5, 4))
    plt.imshow(img); plt.axis("off")
    plt.title(f"{binary}\nType: {pred_class}  |  Tumor prob: {tumor_prob:.2%}")
    plt.tight_layout(); plt.show()
    return pred_class, probs

# Example:
# predict("/content/data/Testing/glioma/Te-gl_0010.jpg")
```
